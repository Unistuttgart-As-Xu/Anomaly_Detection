```python
Total_Number_ship = 15
Diff_angle = 20
Diff_speed = 2
Number_period_list = []
for Number_ship in range(1, Total_Number_ship+1):
    Data_name = ("../Input_2/Generation_data/Test_data_2/generation_data_test_%d.csv" %(Number_ship))
    random_float_list = []
    random_float_list1 = []
    random_float_list2 = []
    random_float_list3 = []
    random_float_list4 = []
    random_float_list5 = []
    random_float_list6 = []
    random_float_list7 = []
    random_float_list8 = []
    random_float_list9 = []
    random_float_list10 = []
    COG = []
    
    for i in range(0, 200):
         x1 = round(random.uniform(0,30),1)
         random_float_list.append(x1)

    df1 = pd.DataFrame(random_float_list, 
                      columns=['SOG'], 
                      index=None)

    print (df1)

    direction = [64.7236946499076, 24.647200290271, -8.910599692160437, -101.32674737075915, -85.74125731076312]

    for i in range (0, 1000):
        bearing = np.random.uniform(direction[0]-Diff_angle,direction[0]+Diff_angle)
        if i == 0:
            x = 49.29910
            y = -123.46484
        else :
            origin = geopy.Point(x, y)
            destination = geodesic(kilometers=random_float_list[i-1]*1.852/60.0).destination(origin, bearing)
            x, y = destination.latitude, destination.longitude
            print(x)
            print(y)
            print(bearing)
        if x > 49.34139 or y > -123.32751:
            break

        random_float_list1.append(x)
        random_float_list2.append(y)
        COG.append(bearing)

    df2 = pd.DataFrame(random_float_list1, 
                      columns=['LAT'],       
                      index=None)
    df3 = pd.DataFrame(random_float_list2, 
                      columns=['LON'], 
                      index=None)
    print (df2)
    print (df3)

    for i in range (0, 1000):
        bearing = np.random.uniform(direction[1]-Diff_angle,direction[1]+Diff_angle)
        if i == 0:
            x = random_float_list1[-1]
            y = random_float_list2[-1]
        else :
            origin = geopy.Point(x, y)
            destination = geodesic(kilometers=random_float_list[i-1]*1.852/60.0).destination(origin, bearing)
            x, y = destination.latitude, destination.longitude
            print(x)
            print(y)
            print(bearing)
        if x > 49.40332 or y > -123.28838:
            break

        random_float_list3.append(x)
        random_float_list4.append(y)
        COG.append(bearing)

    random_float_list3 = random_float_list1 + random_float_list3
    random_float_list4 = random_float_list2 + random_float_list4

    df4 = pd.DataFrame(random_float_list3, 
                      columns=['LAT'], 
                      index=None)
    df5 = pd.DataFrame(random_float_list4, 
                      columns=['LON'], 
                      index=None)

    print (df4)
    print (df5)

    for i in range (0, 1000):
        bearing = np.random.uniform(direction[2]-Diff_angle,direction[2]+Diff_angle)
        if i == 0:
            x = random_float_list3[-1]
            y = random_float_list4[-1]
        else :
            origin = geopy.Point(x, y)
            destination = geodesic(kilometers=random_float_list[i-1]*1.852/60.0).destination(origin, bearing)
            x, y = destination.latitude, destination.longitude
            print(x)
            print(y)
            print(bearing)
        if x > 49.43793 or y < -123.30829:
            break

        random_float_list5.append(x)
        random_float_list6.append(y)
        COG.append(bearing)

    random_float_list5 = random_float_list3 + random_float_list5
    random_float_list6 = random_float_list4 + random_float_list6

    df6 = pd.DataFrame(random_float_list5, 
                      columns=['LAT'], 
                      index=None)
    df7 = pd.DataFrame(random_float_list6, 
                      columns=['LON'], 
                      index=None)

    print (df6)
    print (df7)

    for i in range (0, 1000):
        bearing = np.random.uniform(direction[3]-Diff_angle,direction[3]+Diff_angle)
        if i == 0:
            x = random_float_list5[-1]
            y = random_float_list6[-1]
        else :
            origin = geopy.Point(x, y)
            destination = geodesic(kilometers=random_float_list[i-1]*1.852/60.0).destination(origin, bearing)
            x, y = destination.latitude, destination.longitude
            print(x)
            print(y)
            print(bearing)
        if x < 49.41985 or y < -123.39721:
            break

        random_float_list7.append(x)
        random_float_list8.append(y)
        COG.append(bearing)

    random_float_list7 = random_float_list5 + random_float_list7
    random_float_list8 = random_float_list6 + random_float_list8

    df8 = pd.DataFrame(random_float_list7, 
                      columns=['LAT'], 
                      index=None)
    df9 = pd.DataFrame(random_float_list8, 
                      columns=['LON'], 
                      index=None)

    print (df8)
    print (df9)

    for i in range (0, 1000):
        bearing = np.random.uniform(direction[4]-Diff_angle,direction[4]+Diff_angle)
        if i == 0:
            x = random_float_list7[-1]
            y = random_float_list8[-1]
        else :
            origin = geopy.Point(x, y)
            destination = geodesic(kilometers=random_float_list[i-1]*1.852/60.0).destination(origin, bearing)
            x, y = destination.latitude, destination.longitude
            print(x)
            print(y)
            print(bearing)
        if x > 49.42409 or y <  -123.46381:
            break

        random_float_list9.append(x)
        random_float_list10.append(y)
        COG.append(bearing)

    random_float_list9 = random_float_list7 + random_float_list9
    random_float_list10 = random_float_list8 + random_float_list10


    df10 = pd.DataFrame(random_float_list9, 
                      columns=['LAT'], 
                      index=None)
    df11 = pd.DataFrame(random_float_list10, 
                      columns=['LON'], 
                      index=None)
    for i in range(len(COG)):
      if COG[i] >= 0 and COG[i] < 90:
        COG[i] = COG[i] + 270
      elif COG[i] >= 90 and COG[i] < 180:
        COG[i] = COG[i] - 90
      elif COG[i] >= -90 and COG[i] < 0:
        COG[i] = COG[i] + 90
      elif COG[i] >= -180 and COG[i] < -90:
        COG[i] = COG[i] + 450
    df_COG = pd.DataFrame(COG, 
                columns=['COG'], 
                index=None)

    print (df10)
    print (df11)
    print (df_COG)

    df =pd.concat([df10,df11,df1,df_COG],axis=1, join='inner')
    print(df)

    df.to_csv(Data_name)
    # segment every 10 min of one ship in a file
    Number_period = int(len(COG) / 10) + 1
    Number_period_list.append(Number_period)
    for i in range(0, Number_period):
      fill_zero = random.randint(0,5)
      segment_name_fillzero = ("../Input_2/Generation_data/Test_data_2/generation_data_fillzero_test_ship_%d_period_%d.csv" %(Number_ship,i))
      df_period_fillzero = df.iloc[i*10:(i+1)*10-fill_zero]
      df_period_fillzero.to_csv(segment_name_fillzero)

      segment_name = ("../Input_2/Generation_data/Test_data_2/generation_data_test_ship_%d_period_%d.csv" %(Number_ship,i))
      df_period = df.iloc[i*10:(i+1)*10]
      df_period.to_csv(segment_name)


np.save("../Input_2/Generation_data/Test_data_2/Number_period_list_test.npy", Number_period_list)
```
