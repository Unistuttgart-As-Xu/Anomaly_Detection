```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
train_data = pd.read_csv('final_svm_train.csv')
train_data = shuffle(train_data)
```

```python
train_data.head()
```

```python
len(train_data)
```

```python
X = train_data.drop('label',axis = 1)
X = train_data[["LAT","LON","SOG","COG"]]
y = train_data['label'].values.reshape(-1,1)## use minmax scaler
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
Xtr_s = StandardScaler()
X_train = Xtr_s.fit_transform(X_train)
Xte_s = StandardScaler()
X_test = Xte_s.fit_transform(X_test)
ytr_s  = MinMaxScaler()
y_train = ytr_s.fit_transform(y_train)
yte_s = MinMaxScaler()
y_test = yte_s.fit_transform(y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.svm import SVC
```

```python
svc = SVC(C=1.0, break_ties=False, cache_size=200, class_weight=None, coef0=0.0,decision_function_shape='ovr', degree=3, gamma='scale', kernel='rbf',max_iter=-1, probability=False, random_state=None, shrinking=True,tol=0.001, verbose=False)
```

```python
svc.fit(X_train,y_train)
```

```python
pre= svc.predict(X_test)
```

```python
print("Accuracy Score is  {}".format(accuracy_score(y_test,pre)))
print(confusion_matrix(y_test,pre))
print(classification_report(y_test,pre))
##svm 分类精度1.0
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,
                   multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=None, solver='lbfgs', tol=0.0001, verbose=0,
                   warm_start=False)
```

```python
LR.fit(X_train,y_train)
```

```python
y_pred = LR.predict(X_test)
```

```python
print("Accuracy Score is  {}".format(accuracy_score(y_test,y_pred)))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
